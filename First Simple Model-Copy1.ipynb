```python
df = pd.read_csv("Redfin Data\Redfin_98103-1_Descriptions.csv")
```

```python
df.head()
```

```python
df = df.dropna(subset=["Description"])
```

```python
df['Description'] = df['Description'].astype(str)
```

```python
df["Description"] = df["Description"].str.lower()
```

```python
df["Description"] = df["Description"].apply(lambda x: word_tokenize(x))
```

```python
stop_words = set(stopwords.words('english'))
```

```python
cleaned_tokenized_descriptions = []
for tokens in df["Description"]:
    cleaned_tokens = [word for word in tokens if word.lower() not in stop_words ]
    cleaned_tokenized_descriptions.append(cleaned_tokens)
```

```python
df["Description"] = cleaned_tokenized_descriptions
```

```python
df['Description'] = df['Description'].apply(lambda tokens: [word for word in tokens if word.isalnum()])

```

```python
descriptions = df["Description"]
```

```python
df["Description"]
```

```python
df["Description"]
```

```python
df['Description'] = df['Description'].apply(lambda words: ' '.join(words))
```

```python
df
```

```python
tfidf_vectorizer = TfidfVectorizer(min_df = .01, max_df=.99)
tfidf = tfidf_vectorizer.fit_transform(df["Description"])
tfidf_df = pd.DataFrame(tfidf.toarray(), columns=tfidf_vectorizer.get_feature_names())
```

```python
tfidf_df
```

```python
from gensim import matutils
from gensim.corpora import Dictionary

# Create a Gensim dictionary
gensim_dict = Dictionary(descriptions)


# Convert the TF-IDF matrix to a Gensim corpus
gensim_corpus = matutils.Sparse2Corpus(tfidf.transpose())

```

```python
from gensim.models import LdaModel

num_topics = 10  # Specify the number of topics
lda_model = LdaModel(gensim_corpus, num_topics=num_topics, id2word=gensim_dict, passes=10)
```

```python
topics = lda_model.print_topics(num_topics=num_topics)
```

```python
for topic in topics:
    print(topic)
```
